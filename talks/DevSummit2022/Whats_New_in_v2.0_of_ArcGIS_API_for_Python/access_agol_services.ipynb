### Get a list of AGOLServerManager objects from the servers manager  - _list()_

```python
# list() returns instances of the AGOLServerManager class
for server in ago_servers.list():
    print(f"{type(server)}")
```

```python
# properties and methods of the AGOLServer Manager objects
for server in ago_servers.list():
    print(f"{[d for d in dir(server) if not d.startswith('_')]}")
```

```python
# determine which server(s) is/are tile servers
print(f"Server URL{' '*72}Tile Server\n{'-'*75}{' '*7}{'-'*11}")
for server in ago_servers.list():
    print(f"{server._url:75}{' '*7}{server.is_tile_server}")
```
