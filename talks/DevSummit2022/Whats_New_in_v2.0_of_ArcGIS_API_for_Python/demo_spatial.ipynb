# Spatially Enabled DataFrame

```python
import pandas as pd
from arcgis.gis import GIS
```

```python
gis = GIS(profile="your_online_profile")
```

```python
cone_item = gis.content.get("06f26d149448425e825490e1fcbda1d3")
cone_lyr = cone_item.layers[0]
```

```python
# Query a layer and return dataframe with a SHAPE column
cone_df = cone_lyr.query(as_df=True)
```

```python
# Can now use the spatial property
type(cone_df.spatial)
```
